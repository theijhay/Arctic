```python
#  Preprocess Text Data

import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Remove non-alphabetic characters
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    # Convert to lowercase
    text = text.lower()
    # Tokenize
    words = text.split()
    # Remove stopwords and lemmatize
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    # Join words back into a single string
    return ' '.join(words)

# Apply preprocessing to TITLE and DESCRIPTION
df['clean_title'] = df['TITLE'].apply(preprocess_text)
df['clean_description'] = df['DESCRIPTION'].apply(preprocess_text)
```

```python
# Function to balance the dataset
def balance_dataset(df, target_column, max_samples=100):
    category_counts = df[target_column].value_counts()
    min_count = min(max_samples, category_counts.min())
    
    balanced_df = pd.DataFrame()
    
    for category in category_counts.index:
        category_df = df[df[target_column] == category]
        if len(category_df) > min_count:
            balanced_category_df = resample(category_df, replace=False, n_samples=min_count, random_state=42)
        else:
            balanced_category_df = category_df
        balanced_df = pd.concat([balanced_df, balanced_category_df])
    
    return balanced_df

# Balance the dataset for categorization
balanced_df = balance_dataset(df, 'clean_categories', max_samples=100)

# Verify the balancing
print(balanced_df['clean_categories'].value_counts())

```
